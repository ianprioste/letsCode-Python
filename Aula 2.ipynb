## Enquanto (while)

Melhor para validação de entradas

### Loop Infinito

```python
while True:
    nome = input ("Digite o seu nome: ")
    nome = nome.upper().capitalize()
        
    if nome == "Parar":
        break
    else: 
        print ("oi", nome)
```

### Loop condicional

```python
# Loop condicional

nome = ""
while nome != "Parar":
    nome = input ("Digite o seu nome: ")
    nome = nome.upper().capitalize()
    print ("oi", nome)
```

# Para cada (FOR) 

Melhor para contadores e trabalhar com listas.

### Criando Intervalo

```python
print (list(range(0,10,1)))
```

### Estrutura do FOR

```python
# Não precisa inicializar o i 

for i in range (2):
    print (i)
```

### Lista

```python
notas = [0, 3, 5, 9]
print(notas)

#Primeiro
print(notas[0])

#Último
print(notas[-1])
```

### Lista com 'while'

```python
notas = [0, 3, 5, 9]
print(notas)

i = 0
while i < len(notas):
    print (notas[i])
    i+=1
```

### Lista com 'for'

```python
# Quando o indice será utilizado

for i in range (len(notas)):
    print(notas[i])
```

```python
# Quando o indice não será utilizado

for nota in notas:
    print(nota)
```

```python
# Quando o indice e o valor será utilizado
for i, nota in enumerate (notas):
    print("Nota "+ str(i)+":"+str(nota))
    
```

### Interação de duas listas

```python
lista1 = [0,1,2,3]
lista2 = [-1,6,2,1]

for i in range(len(lista1)):
    print(lista1[i]+lista2[i])
```

### Cópia

```python
#cópia rasa
lista1 = [0,1,2,3]
lista2 = lista1

print (lista1,lista2)

lista1[0]=5
lista2[3]=4
print(lista1,lista2)

#Cópia - Deixar as listas independentes
lista2=lista1.copy()
lista1[0]=0
lista2[3]=12

print (lista1, lista2)

#Cópia criando uma nova lista
lista = [2,3,5,7,11,19,20,34,23]
lista1 = lista[1:5]
lista2 = lista[:5]
lista3 = lista[6:]
lista4 = lista[:]
lista5 = lista[0:10:2]

print (lista1)
print (lista2)
print (lista3)
print (lista4)
print (lista5)
```

#### Trabalhando com Matriz

```python
matriz = [[0,1,2],[3,4,5]]

for linha in matriz:
    for elemento in linha:
        print(elemento, end=" " )
    print()
```
