## Definir função

```python
def hello():
    print ("Olá")
```

```python
hello()
```

```python
# def nomedafuncao (parametros)
def hello2(nome):
    print ("Olá",nome)
```

```python
#nomedafuncao(argumentos)

hello2("Argumento")
hello2("Victor")
hello2("João")
hello2("Ian")
```

```python
def soma(a,b):
    print(a+b)
```

```python
soma(2,3)
soma(3,4)
x = soma(1,2)
print("x=",x)
```

```python
def subtracao(a,b):
    print (a-b)
```

```python
#a ordem dos argumentos importa para a função
subtracao(2,3)
subtracao(3,2)
```

```python
def soma(a,b=0):
    return(a+b)
```

```python
x = soma(1,2)
print (x)

y = soma(1)
print(y)
```

```python
def fatorial (n):
    if n<0:
        return -1
    
    fat = 1
    while n > 1:
        fat*=n
        n-=1
    return fat

print(fatorial(6))
```

```python
#O nome da varíavel fora da função não precisa ser o mesmo nome de dentro da função.
#As variáveis criadas dentro da função só existem dentro da mesma.

def func(x):
    x+=10

x=5
func(x)
print (x)

def func(x):
    x+=10
    return (x)

x = func(5)
print(x)
```
